##Part1: Removing Missing Values

###1 How do you use Pandas to read the company_emp1.csv file into a DataFrame?

```python
import pandas as pd
df_company_emp1 =pd.read_csv("/content/drive/MyDrive/Phase2/Week2/Day3/Copy of company_emp1.csv")
df_company_emp1
```

###2 What code would you use to print the number of missing values in each column?

```python
df_company_emp1.isnull().sum()
```

###3 How can you calculate and print the total number of missing values in the entire DataFrame?

```python
df_company_emp1.isnull().sum().sum()
```

###4 How can you keep only the rows with at least 3 non-NA values?

```python
df_company_emp1.dropna(thresh=3,inplace=True)
df_company_emp1
```

###5 What code would you use to drop rows that have missing values in the 'buyer' column?

```python
df_company_emp1.dropna(subset=['buyer'], inplace = True)
df_company_emp1.reset_index(inplace=True)
df_company_emp1.isnull().sum()
```

###6 How do you drop rows that have all values missing?

```python
df_company_emp1.dropna(how='all')
df_company_emp1.isnull().sum()
```

###7 What method can you use to drop all rows that contain any missing values?

```python
df_company_emp1.dropna(inplace=True)
df_company_emp1.reset_index(inplace=True, drop=True)
df_company_emp1.isnull().sum()
```

##Part2: Replacing Missing Values

###8 Use Pandas to read company_emp2.csv.

```python
df_company_emp2 =pd.read_csv("/content/drive/MyDrive/Phase2/Week2/Day3/Copy of company_emp2.csv")
df_company_emp2.isnull().sum()
```

###9 What command replaces missing values in the time column with the value 60?

```python
df_company_emp2["time"].fillna(60, inplace = True)
df_company_emp2.isnull().sum()
```

###10 What code replaces missing values in the sales column with the mean of that column?

```python
df_company_emp2['sales'].fillna(df_company_emp2['sales'].mean(), inplace=True)
df_company_emp2.isnull().sum()
```

###11 What command replaces missing values in the visitor column with the maximum value of that column?

```python
df_company_emp2['visitor'].fillna(df_company_emp2['visitor'].max(), inplace=True)
df_company_emp2.isnull().sum()
```

##Part3:Formatting Issues

###12 Use Pandas to read company_format1.csv.

```python
df_company_format1 =pd.read_csv("/content/drive/MyDrive/Phase2/Week2/Day3/Copy of company_format1.csv")
df_company_format1.head(10)
```

###13 Convert companyid to int

```python
df_company_format1['companyid'] = df_company_format1['companyid'].astype(int)
```

###14 Convert Date to datetime

```python
df_company_format1.iloc[5, 2] = "20/08/2018"
df_company_format1.iloc[17, 2] = "20/08/2018"
print(df_company_format1['Date'].dtype)
df_company_format1['Date'] = pd.to_datetime(df_company_format1['Date'], format='%d/%m/%Y')
print(df_company_format1['Date'].dtype)

```

###15 Convert buyer to int by removing characters.

```python
df_company_format1['buyer'] = df_company_format1['buyer'].str.replace('\D', '', regex=True).astype('int')
df_company_format1.head(10)
```

###16 Convert visitor to int and replace missing values with 180.

```python
df_company_format1['visitor'] = df_company_format1['visitor'].str.replace('\D', '', regex=True).replace('', pd.NA) .fillna(180).astype('int')
```
